CORELATION MATRIX

```python
data.columns = data.columns.str.strip()
for col in data.columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')

data = data.dropna(axis=1, how='all')
data = data.dropna(axis=0, how='any')

correlation_matrix = data.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=False, fmt=".2f", cmap='coolwarm', square=True,
            cbar_kws={"shrink": .8}, linewidths=0.5)

plt.title('Correlation Matrix', fontsize=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
print("\nCorrelation Matrix:\n", correlation_matrix)
```

```python
columns_to_drop = ['event_number', 'Time', 'file_number']
data = data.drop(columns=columns_to_drop, errors='ignore')
print("Updated columns in the dataset:", data.columns.tolist())
```

KURTOSIS

```python
kurtosis_values = data.kurtosis()
print("Kurtosis of features:")
print(kurtosis_values)
```

```python
fig, axes = plt.subplots(6, 7, figsize=(20, 20))
axes = axes.flatten()

for i, feature in enumerate(kurtosis_values.index):
    sns.kdeplot(data[feature], color='blue', fill=True, ax=axes[i])
    axes[i].set_title(f'{feature} (Kurtosis: {kurtosis_values[feature]:.2f})', fontsize=10)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Density')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

SKEWNESS

```python
skewness_values = data.skew()
print("Skewness of features:")
print(skewness_values)
```

```python
fig, axes = plt.subplots(6, 7, figsize=(20, 20))
axes = axes.flatten()
for i, feature in enumerate(skewness_values.index):
    sns.histplot(data[feature], bins=30, kde=True, color='skyblue', ax=axes[i])
    axes[i].set_title(f'{feature} (Skewness: {skewness_values[feature]:.2f})', fontsize=10)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

X = data.drop(columns=['label'])
y = data['label']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


pca = PCA(n_components=20)
X_pca = pca.fit_transform(X_scaled)

explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)
std_deviation = np.sqrt(pca.explained_variance_)
pca_summary = pd.DataFrame({
    'Standard Deviation': std_deviation,
    'Proportion of Variance': explained_variance_ratio,
    'Cumulative Proportion': cumulative_variance
}, index=[f'PC{i+1}' for i in range(len(std_deviation))])

print(pca_summary)
```

```python
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Label')
plt.title('PCA Scatter Plot')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Plotting in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=data['label'], cmap='viridis', alpha=0.7)

plt.colorbar(scatter, label='Label')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA Visualization')

plt.show()
```
